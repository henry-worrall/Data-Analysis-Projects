# Exchange Rates

This notebook uses explanatory data visualisation to tell a story with [data](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020) from European Central Bank. The aim of this project is to show how we can use Matplotlib to create easy to understand visualisations. Hopefully, the final visualisation should show what is trying to be done with the data!

```python
# load packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load data
exchange_rates = pd.read_csv('Data/euro-daily-hist_1999_2020.csv')
exchange_rates.head()
```

## Data Cleaning

Our focus in this project will be on the exchange rate between the euro and the American dollar. Therefore, I shall rename the columns to something easier to handle.

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', ignore_index=True,
                           inplace=True)
```

```python
exchange_rates.head()
```

```python
# Isolate time and US Dollar
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()
```

```python
# Drop rows with '-' in
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar['US_dollar']
```

## Visualisation

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

```python
# Calculate a rolling mean
euro_to_dollar['rolling mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
# plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling mean'], color='red', linewidth=2)
plt.show()
```

## Creating a Story

Let's try and have a look at how the exchange rate has changed over the course of the pandemic. To do this, we can highlight the pandemic as the area of interest by making the line stand out, and enclosing the area. We can also filter out the years before 2015.

```python
# filter data to be after 2015
post_2015 = euro_to_dollar[euro_to_dollar['Time'].dt.year>=2015]
post_2015
```

```python
import datetime as dt
# segment data for pandemic
pandemic = post_2015[post_2015['Time'].dt.date > dt.date(2020,3,10)]
pandemic
```

```python
import matplotlib.ticker as ticker
# create an object orientated interface plot
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(12,6))

# plot line with post_2015 data
ax.plot(post_2015['Time'], post_2015['rolling mean'],
        color='#b00b1e', alpha=0.2)

# plot line with pandemic data
ax.plot(pandemic['Time'], pandemic['rolling mean'],
        color='#b00b1e', alpha=1)

# Highlihting the change during the pandemic in the background
ax.axvspan(xmin=dt.date(2020,3,10), xmax=dt.date(2021,1,1),
           alpha=0.3, color='grey')

# remove spines
for location in ['top','left','bottom','right']:
    ax.spines[location].set_visible(False)

# ajust y labels
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.05))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.01))

# reduce tick colour
ax.tick_params(axis='y', colors='grey')
ax.tick_params(axis='x', colors='grey')

# add title
# ax.text(13020, 1.67, "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",
#         weight='bold')
# ax.text(13020, 1.07, s='The Euro-USD exchange rate increased by over 10 during the pandemic')
ax.text(dt.date(2014,5,1),1.25,'Euro-USD exchange rates between 2015-2021')
ax.text(dt.date(2014,5,1),1.263,
            'The Euro has increased 10% in value against the dollar since the start of the pandemic',
            weight='bold', size=16)

# Adding a signature
ax.text(dt.date(2014,5,1), 1.02, 'Â©Henry Worrall' + ' '*103 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=14)



plt.show()
```

```python
# compare exchange rate for presidents 
# Bush 2001,1,20 to 2009,1,20
# obama - 2009,1,20 to 2017,1,20
# Trump 2017,1,20 to 2021,1,20

bush_years = euro_to_dollar[euro_to_dollar['Time'].dt.date.between(dt.date(2001,1,20),dt.date(2009,1,20))]
obama_years = euro_to_dollar[euro_to_dollar['Time'].dt.date.between(dt.date(2009,1,20),dt.date(2017,1,20))]
trump_years = euro_to_dollar[euro_to_dollar['Time'].dt.date.between(dt.date(2017,1,20),dt.date(2021,1,20))]

bush_years
```
